```python
# year value check non value and cleaning non integer data

car=car[car['year'].str.isnumeric()]
```

```python
#  yaer datatype change object to int

car['year']=car['year'].astype(int)
```

```python
# Remove "ask for price " tag from price faeture

car=car[car['Price']!="Ask For Price"]
```

```python
car.info()
```

```python
# Remove , coma and chnage data type object ot int 

car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

```python
# Remove kms Tag and coma , from kms driven feature

car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
# check which values is numeric on kms driven

car=car[car['kms_driven'].str.isnumeric()]
```

```python
car.info()
```

```python
# change data type of kms driven object into int

car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car.info()
```

```python
# Remove nan value from fuel type 

car=car[~car['fuel_type'].isna()]
```

```python
# take only frist three string with space

car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car=car.reset_index(drop=True)
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

```python
X=car.drop(columns='Price')
y=car['Price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
score=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(score))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)   
    
```
