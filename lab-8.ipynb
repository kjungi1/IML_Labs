```python
# (8.1) Write a Pandas program to create a line plot of the opening, closing stock prices of given company
# between two specific dates

# KISHAN JUNGI
# 216270307014

import pandas as pd
import matplotlib.pyplot as plt

# Load your stock price data into a Pandas DataFrame (replace 'your_data.csv' with your data source)
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/BRITANNIA.csv')

# Assuming your dataset has a column 'Date' that contains date information, make sure to set it as the index.
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Define the specific date range
start_date = '2000-01-03'
end_date = '2000-03-13'

# Filter the data for the specified date range
filtered_data = df.loc[start_date:end_date]

# Create a line plot of the opening and closing prices
plt.figure(figsize=(12, 6))
plt.plot(filtered_data.index, filtered_data['Open'], label='Opening Price', color='blue')
plt.plot(filtered_data.index, filtered_data['Close'], label='Closing Price', color='red')

plt.title('Opening and Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```
